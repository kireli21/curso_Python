```python
#Programa que crea un archivo para escritura

n1 = "Edwin"
n2 = "Pablo"
n3 = "Ruben"
c1 = 10
c2 = 9
c3 = 8
salida = open("prueba.txt", "w") #csv para exel
for c in range(1):
  salida.write(n1 + " " + str(c1) + "\n")
  salida.write(n2 + " " + str(c2) + "\n")
  salida.write(n3 + " " + str(c3) + "\n")
salida.close()
```

```python
class Persona:
  nombre = ""
  calif = ""

  def __init__ (self, n, c):
    self.nombre = n
    self.calif = c

listaper = []
listaper.append(Persona("Pedro Paramo", 9))
listaper.append(Persona("Pedro Marmol", 9))
listaper.append(Persona("Pedro Reyes", 6))
listaper.append(Persona("Pedro Dias", 9))
listaper.append(Persona("Pedro Cardo", 5))
listaper.append(Persona("Pedro Mercado", 8))
listaper.append(Persona("Pedro Piña", 4))

salida = open("prueba.txt", "w")
for p in listaper:
  salida.write(p.nombre + ", " + str(p.calif) + "\n")
salida.close()
```

```python
#Par -> lo dividimos a la mitad
#impar -> lo *3 y le +1
import matplotlib.pyplot as plt

def esPar(n):
  return (n%2) == 0

def ulam(n):
  if esPar(n):
    return n//2
  else:
    return (n*3)+1

def ruta(n):
  r = ""
  while (n != 1):
    r = r + str(n) + "-"
    n = ulam(n)
  r=r + "1"
  return(r)

lista = []
for c in range(1,200):
    lista.append(ruta(c).count("-"))

plt.plot(lista)
plt.show()
```
